```python
import pandas as pd
import numpy as np
import re
```

```python
temp_total = pd.read_csv('(최종)모든자료date_ngram_label.csv', encoding='utf-8', index_col=0)
```

```python
temp_total['ngram'] = temp_total['ngram'].astype('str')
```

```python
temp_total['ngram'] = temp_total['ngram'].apply(lambda x:re.sub('[가-힣]{1,8}\/(JKS|JKC|JKG|JKO|JKB|JKV|JKQ|JC|JX|EP|EF|EC|ETN|ETM|XPN|XSN|XSV|XSA|XR|SF|SE|SSO|SSC|SC|SY|SH|SL|SN)(\,?)','',str(x),50))
temp_total['ngram'] = temp_total['ngram'].apply(lambda x: re.sub('\s+[가-힣]{1,8}\/(JKS|JKC|JKG|JKO|JKB|JKV|JKQ|JC|JX|EP|EF|EC|ETN|ETM|XPN|XSN|XSV|XSA|XR|SF|SE|SSO|SSC|SC|SY|SH|SL|SN)(\S*|\,)','',x,50))
```

```python
temp_total
```

```python
## df_all 만들기: 전체 ngram 카운팅
temp_ls = [] # 빈리스트 생성
label_ls = []
for i in range(len(temp_total)): # 사용할 데이터프레임의 길이로 for문 돌리기
    try:
        test = temp_total['ngram'][i].replace(' ','').split(',') # 컬럼이름 ngram의 i 번째를 split
        label = temp_total['updown'][i]
        temp_ls.extend(test) # 빈 리스트에 한번에 추가
        label_ls.extend([label]*len(test))
        if i%10000 == 0 : # %: 나머지, 1000개마다 프린트
            print('{}/{}'.format(i, temp_total.shape[0]))
    except Exception as ex: # 에러나면 에러 이름
        print("============== {} : {} ==============".format(i, str(ex)))
#         print(row[‘content’])
        pass
temp_df = pd.DataFrame({'ngram':temp_ls,'updown':label_ls}, columns=['ngram', 'updown']) # 새로운 데이터프레임 생성
#df_all = pd.DataFrame(temp_df[‘ngram’].value_counts()) # value_counts사용
#df_all = df_all.reset_index()
#df_all.head(3) (edited) 
```
