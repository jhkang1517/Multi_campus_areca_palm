#### 카드 txt 파일 불러오기

```python
# 참조 https://pinkwink.kr/958

card_raw = pd.read_csv('CARD_SPENDING_190809.txt', sep='\t', header = 0, encoding= 'utf-8', names = ['STD_DD', 'GU_CD','DONG_CD','MCT_CAT_CD','SEX_CD','AGE_CE','USE_CNT','USE_AMT'] )
```

#### 컬럼명 변경

```python
card.columns = ['기준일자','구코드','행정동코드','업종코드','성별코드','나이코드','이용건수','이용금액']
```

#### 원본 보호

```python
card = card_raw.copy()
```

```python
card['기준일자'] = pd.to_datetime(card['기준일자'])
```

### 랜덤 샘플 1,500,000개 층화 추출

sklearn 참고 : https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedShuffleSplit.html

랜덤샘플 참고: 
https://cnpnote.tistory.com/entry/PYTHON-Pandas-%EB%8D%B0%EC%9D%B4%ED%84%B0-%ED%94%84%EB%A0%88%EC%9E%84%EC%9D%98-%EB%AC%B4%EC%9E%91%EC%9C%84-%ED%96%89-%EC%84%A0%ED%83%9D

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits = 10, test_size = 0.75, random_state=42)

for train_index, test_index in split.split(card, card['행정동코드']):
    strat_train_set = card.loc[train_index]
    strat_test_set = card.loc[test_index]
```

```python
strat_test_set['행정동코드'].value_counts()
```

```python
card['행정동코드'].value_counts()
```

```python
card_sample = strat_test_set
```
