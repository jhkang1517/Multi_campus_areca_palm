# 2. 통합 ngram (2~5gram + tokens) 추출하기

### return 값을 2개 반환하여 ngram과 tokens를 동시에 받아온다

```python
import pandas as pd
text_yh_0507 = pd.read_csv('연합뉴스 2005-2017 전처리 끝_190717_집에서.csv', encoding='utf-8', index_col=0)
text_yh_0507 = text_yh_0507.astype('str')

# n-gram 코드

import pandas as pd
from ekonlpy.sentiment import MPCK


# return을 ngram과 token 각각 받는다
def text2ngram(text):
    mpck = MPCK()
    bef_tokens = mpck.tokenize(text)
    ngrams = mpck.ngramize(bef_tokens)
    return ngrams, bef_tokens             
```

### ngram과 token을 각각 리스트에 임시 저장한다

token리스트 & ngram 리스트

시간이 아주 많이 걸리므로 돌려놓고 잠깐 쉬도록 하자

```python



test_tokens = []
test_ngrams = []

for i in range(len(text_yh_0507)):
    a, b = text2ngram(text_yh_0507['content'][i])
    test_ngrams.append(a)
    test_tokens.append(b)
```

```python
# ngram과 token을 하나의 리스트로 합친다.
# 시간이 아주 많이 걸리므로 돌려놓고 잠깐 쉬도록 하자

test_total =[]
for i in range(len(test_ngrams)):
    test_total.append(test_ngrams[i]+test_tokens[i])
    
len(test_total)
```

#### 합친 리스트를 ngram이라는 컬럼으로 만든다

```python
# 

text_yh_0507['ngram'] = test_total
```

#### ngram 정제작업을 거친다. 

#### 작은 따옴표는 2000정도로 충분히 제거해야한다.

```python

import re

text_yh_0507 = text_yh_0507.astype('str')

text_yh_0507['ngram'] = text_yh_0507['ngram'].apply(lambda x: re.sub("'",'',x,2000))
text_yh_0507['ngram'] = text_yh_0507['ngram'].apply(lambda x: re.sub("\[",'',x,10))
text_yh_0507['ngram'] = text_yh_0507['ngram'].apply(lambda x: re.sub("\]",'',x,10))


```
